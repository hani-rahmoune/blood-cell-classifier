```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("masoudnickparvar/white-blood-cells-dataset")

print("Path to dataset files:", path)
```

```python
TRAIN_PATH = path + "/Train"
TEST_A_PATH = path + "/Test-A"
TEST_B_PATH = path + "/Test-B"
```

```python
print(TRAIN_PATH)
print(TEST_A_PATH)
print(TEST_B_PATH)
```

```python
import random , pandas as pd , os
```

```python
random.seed(42) #same split each time
```

```python
print(os.listdir(TRAIN_PATH))
```

```python
classes = sorted(os.listdir(TRAIN_PATH)) #returns the name of the files and directories that fall 1 step under the path arg
print(classes)
```

```python
class_to_index = dict((c, i) for i, c in enumerate(classes))
print(class_to_index)
```

```python
x = os.path.join(TRAIN_PATH, classes[0])
print(x)
```

```python
def create_path_list(path,name):
  path_list = []
  for c in classes:
    for file_name in os.listdir(os.path.join(path, c)):
        if file_name.lower().endswith((".jpg",".jpeg",".png")):
          path_list.append((f'{name}/{c}/{file_name}',class_to_index[c]))
  return path_list
```

```python
TRAIN_path_list = create_path_list(TRAIN_PATH,"Train")
TEST_A_path_list = create_path_list(TEST_A_PATH,"Test-A")
```

```python
classes_B = sorted(os.listdir(TEST_B_PATH))
class_to_index_B = dict((c, i) for i, c in enumerate(classes_B))
TEST_B_path_list = []
for c in classes_B:
  for file_name in os.listdir(os.path.join(TEST_B_PATH, c)):
      if file_name.lower().endswith((".jpg",".jpeg",".png")):
        TEST_B_path_list.append((f'"Test_B"/{c}/{file_name}',class_to_index_B[c]))
```

```python
print(TEST_B_path_list)
```

```python
random.shuffle(TRAIN_path_list)
random.shuffle(TEST_A_path_list)
random.shuffle(TEST_B_path_list)
```

```python
from sklearn.model_selection import train_test_split

paths = [x[0] for x in TRAIN_path_list]
labels = [x[1] for x in TRAIN_path_list]
print(paths)
print(labels)
```

```python
train_paths, val_paths, train_labels, val_labels=train_test_split(
    paths,
    labels,
    test_size=0.2,
    random_state=42,
    stratify=labels,
)
```

```python
train_data = list(zip(train_paths, train_labels))
val_data = list(zip(val_paths, val_labels))
print(train_data)
print(val_data)
```
